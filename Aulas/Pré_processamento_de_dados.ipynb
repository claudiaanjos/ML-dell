### Transformando a escala dos dados

T√©cnica de pr√©-processamento para alterar a escala dos dados, o intervalo no qual os dados est√° abrangendo.

A ordem de grandenza das colunas √© diferente.

Vamos selecionar apenas as colunas num√©ricas:

```python
data = pd.read_table('fruit_data_with_colors_miss.txt',na_values=['.','?'])
data = data.fillna(data.mean())
data = data[['mass','width','height','color_score']]
```

```python
data
```

Para cada coluna iremos capturar o valor que queremos transformar e subtrair do valor m√≠nimo da coluna dividido pelo valor m√°ximo subra√≠do do m√≠nimo.

(valor - min) / (max - mix)

Para aplicar essa t√©cnica iremos utilizar o sklearn: MinMaxScaler.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

A fun√ß√£o fit √© muito utilizada no aprendizado de m√°quinas.
Essa fun√ß√£o ir√° construir para cada coluna do conjuntos de dados o m√°ximo e min√≠mo e armazenar no mm.

```python
mm.fit(data)
```

Agora vamos retornar os dados transformados.

### Encontrando outliers

Outliers: elemento que possui entre suas caracter√≠sticas estar fora do padr√£o, n√£o s√£o o que acontece normalmente. 
Iremos eliminar esses elementos, pois queremos que o modelo aprenda padr√µes.

```python
data = pd.read_table('fruit_data_with_colors_miss.txt',na_values=['.','?'])
data = data.fillna(data.mean())
```

```python
macas = data[data['fruit_name'] == 'apple']
```

```python
macas.describe()
```

Vamos encontrar outliers pelo desvio padr√£o.

```python
est = macas['mass'].describe()
```

Iremos utilizar a seguinte t√©cnica para verificar outliers:
Queremos as ma√ß√£s as quais a massa √© maior que a m√©dia da coluna de massa das ma√ß√£s + duas vezes o desvio padr√£o.



```python
macas[(macas['mass'] > est['mean'] + (est['std']) * 2)] #acima da m√©dia
```

```python
macas[(macas['mass'] < est['mean'] - (est['std']) * 2)] #abaixo da m√©dia
```
