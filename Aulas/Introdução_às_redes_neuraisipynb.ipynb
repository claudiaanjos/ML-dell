#Implementação de Redes Neurais

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_boston
```

```python
X,y = load_boston(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
#como a rede neural é treinada irá se beneficiar de escalas pequenas e parecidas, por isso usamos o minmax
mm = MinMaxScaler()
X_train = mm.fit_transform(X_train)
X_test = mm.transform(X_test)
```

```python
#ativação como padrão = relu (não permite valores negativos)
#tamanho da camadas (10, 10, 10) = 3 camadas com 10 neurônios cada (podemos ir reajustando)
#Muitas camadas e neurônios = overfitting
mlp = MLPRegressor(hidden_layer_sizes=(100,100,50,50),max_iter=1000)
```

```python
mlp.fit(X_train,y_train)
```

```python
mlp.score(X_test,y_test)
```
